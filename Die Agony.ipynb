```python
import numpy as np
```

```python
class treeNode:
  def __init__(self, xx, yy, mv, topface, frontface, sco):
    self.x = xx
    self.y = yy
    self.move = mv
    self.top_face = topface
    self.front_face = frontface
    self.dice = np.zeros(7)
    self.score = sco
```

```python
move_right = {
    (2,1) : (3,1),
    (3,1) : (5,1),
    (5,1) : (4,1),
    (4,1) : (2,1),
    
    (6,2) : (3,2),
    (3,2) : (1,2),
    (1,2) : (4,2),
    (4,2) : (6,2),
    
    (2,3) : (6,3),
    (6,3) : (5,3),
    (5,3) : (1,3),
    (1,3) : (2,3),

    (2,4) : (1,4),
    (1,4) : (5,4),
    (5,4) : (6,4),
    (6,4) : (2,4),

    (4,5) : (1,5),
    (1,5) : (3,5),
    (3,5) : (6,5),
    (6,5) : (4,5),

    (4,6) : (5,6),
    (5,6) : (3,6),
    (3,6) : (2,6),
    (2,6) : (4,6)
}

move_left = {
    (2,1) : (4,1),
    (4,1) : (5,1),
    (5,1) : (3,1),
    (3,1) : (2,1),
    
    (6,2) : (4,2),
    (4,2) : (1,2),
    (1,2) : (3,2),
    (3,2) : (6,2),
    
    (2,3) : (1,3),
    (1,3) : (5,3),
    (5,3) : (6,3),
    (6,3) : (2,3),

    (2,4) : (6,4),
    (6,4) : (5,4),
    (5,4) : (1,4),
    (1,4) : (2,4),

    (4,5) : (6,5),
    (6,5) : (3,5),
    (3,5) : (1,5),
    (1,5) : (4,5),

    (4,6) : (2,6),
    (2,6) : (3,6),
    (3,6) : (5,6),
    (5,6) : (4,6)
}

def move_up(top, front):
    return front, 7-top

def move_down(top, front):
    return 7-front, top
```

```python
def backtrack(node):
    global visited

    xx = node.x
    yy = node.y
    if grid[xx][yy] == 732:
        print(f'Found: {node.dice}')
        print(f'{xx} , {yy}')
        visited[(xx,yy)] = True
        return True
    
    mv = node.move + 1
    top = node.top_face
    front = node.front_face
    
    #move_left
    if xx>0:
        next_orientation = move_left[(top,front)]
        new_top = next_orientation[0]
        new_front = next_orientation[1]
    
        if node.dice[new_top] == 0:
            newNode = treeNode( xx-1, yy, mv, new_top, new_front, grid[xx-1][yy])

            newNode.dice = np.array(node.dice)
            newNode.dice[new_top] = (grid[xx-1][yy] - node.score) / mv
    
            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True
    
        elif node.score + mv*node.dice[new_top] == grid[xx-1][yy]:
            newNode = treeNode( xx-1, yy, mv, new_top, new_front, grid[xx-1][yy])
    
            newNode.dice = np.array(node.dice)

            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True

    #move_right
    if xx<5:
        next_orientation = move_right[(top,front)]
        new_top = next_orientation[0]
        new_front = next_orientation[1]
    
        if node.dice[new_top] == 0:
            newNode = treeNode( xx+1, yy, mv, new_top, new_front, grid[xx+1][yy])

            newNode.dice = np.array(node.dice)
            newNode.dice[new_top] = (grid[xx+1][yy] - node.score) / mv
    
            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True
    
        elif node.score + mv*node.dice[new_top] == grid[xx+1][yy]:
            newNode = treeNode( xx+1, yy, mv, new_top, new_front, grid[xx+1][yy])
    
            newNode.dice = np.array(node.dice)

            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True

    #move_down
    if yy>0:
        new_top, new_front = move_down(top,front)

        if node.dice[new_top] == 0:
            newNode = treeNode( xx, yy-1, mv, new_top, new_front, grid[xx][yy-1])

            newNode.dice = np.array(node.dice)
            newNode.dice[new_top] = (grid[xx][yy-1] - node.score) / mv
    
            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True
    
        elif node.score + mv*node.dice[new_top] == grid[xx][yy-1]:
            newNode = treeNode( xx, yy-1, mv, new_top, new_front, grid[xx][yy-1])
    
            newNode.dice = np.array(node.dice)

            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True

    #move_up
    if yy<5:
        new_top, new_front = move_up(top,front)
    
        if node.dice[new_top] == 0:
            newNode = treeNode( xx, yy+1, mv, new_top, new_front, grid[xx][yy+1])

            newNode.dice = np.array(node.dice)
            newNode.dice[new_top] = (grid[xx][yy+1] - node.score) / mv
    
            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True
    
        elif node.score + mv*node.dice[new_top] == grid[xx][yy+1]:
            newNode = treeNode( xx, yy+1, mv, new_top, new_front, grid[xx][yy+1])
    
            newNode.dice = np.array(node.dice)

            if backtrack(newNode) == True:
                print(f'{xx} , {yy}')
                visited[(xx,yy)] = True
                return True

    return False  
```

```python
visited = {(i,j) : False for i in range(6) for j in range(6)}
ans = 0
grid = [
    [0, 5, -7, 186, 81, 57] , 
    [77, 23, 2, 42, 123, 33] , 
    [32, -4, 357, 195, 240, 132] , 
    [403, 592, 452, 704, 443, 268] , 
    [337, 445, 317, 452, 353, 492] , 
    [452, 620, 395, 228, 508, 732]
]

first_node = treeNode(0, 0, 0, 2, 1, 0)
if backtrack(first_node) == True:
    print("Son of a bitch!")

for i in range(6):
    for j in range(6):
        if visited[(i,j)] == False:
            ans += grid[i][j]
            print(f'{i},{j} : {grid[i][j]}')
```
