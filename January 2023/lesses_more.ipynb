```python
def check_in_dic(a,b,c,d,table):

    abcd = (a,b,c,d)
    for i in range(4):
        tmp = abcd[i:]
        tmp += abcd[:i]
        if tmp in table:
            return True , tmp

    #reverse
    dcba = (d,c,b,a)
    for i in range(4):
        tmp = dcba[i:]
        tmp += dcba[:i]
        if tmp in table:
            return True , tmp
    
    return False , (-1,-1,-1,-1)
    
def func(a,b,c,d,table):

    exists , tup = check_in_dic(a,b,c,d,table)
    if exists:
        return table[tup]
    
    abcd = (a,b,c,d)
    if a == b == c == d == 0:
        table[abcd] = 1
        return 1
    
    table[abcd] = 1 + func(abs(a-b),abs(b-c),abs(c-d),abs(d-a),table)
    return table[abcd]
```

```python
def calc(init_dim, loop_dim):
    table = {}
    
    max_val = 0.0
    for b in range(loop_dim):
        print(b)
        for c in range(loop_dim):
            for d in range(loop_dim):
                res = func(0,b,c,d,table)
                if res > max_val:
                    max_val = res
                        
    return max_val, table
```
